```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset_train = pd.read_csv('/content/microsoft_stocks.csv')
print('shape is = {}'.format(dataset_train))
print(dataset_train.head())
```

```python
training_set = dataset_train.iloc[:,1:2].values
print('shape is ={}'.format(training_set.shape))
print(training_set[0:5])
```

```python
plt.plot(training_set,color = 'blue',label='Microsoft Stock Price in Test set')
plt.xlabel('Time')
plt.ylabel('Microsoft Stock Price')
plt.legend()
plt.show()
```
